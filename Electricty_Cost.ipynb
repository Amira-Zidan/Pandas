```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'operating_hours': [150, 170, 160, 180, 200, 210], 
    'electricity_cost': [3200, 3500, 3350, 3800, 4100, 4350]  
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df['cost_per_hour'] = df['electricity_cost'] / df['operating_hours']
```

```python
df
```

```python
avg_cost_per_hour = np.mean(df['cost_per_hour'])
avg_cost_per_hour
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[['operating_hours']]
y = df['electricity_cost']
model = LinearRegression(fit_intercept=True)
model.fit(x, y) 
```

```python
model = LinearRegression(
    fit_intercept=True,
    copy_X=True,
    n_jobs=None,
    positive=False
)
```

```python
New_hours_dict = {'operating_hours': [250]}
New_hours_df = pd.DataFrame(New_hours_dict)
prediction_model=model.predict(New_hours_df)
print(f"R2 Score: {model.score(x,y):.4f}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficient: {model.coef_[0]:.2f}")
print(f"Predicted Cost for 250 hours: {prediction_model[0]:.2f}")
```

```python
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel('Operating Hours')
plt.ylabel('Cost')
plt.legend()
plt.show()
```

```python
# 4. Visualization
plt.figure(figsize=(10, 6))

# Plot existing data points
plt.scatter(x, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(x, model.predict(x), color='red', label='Regression Line')

# Plot the new predicted point
plt.scatter(250, prediction_model[0], color='green', marker='x', s=150, label='Forecast (250h)')

# Chart formatting
plt.xlabel('Operating Hours')
plt.ylabel('Electricity Cost ($)')
plt.title('Cost Forecasting Analysis')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 5. Display results
print(f"R2 Score: {model.score(x, y):.4f}")
print(f"Predicted Cost for 250 hours: {prediction_model[0]:.2f}")
```
