```python
import pandas as pd
```

```python
import numpy as np
```

```python
data={
    'Name': ['Ahmed', ' Sara', 'Ahmed', 'Mona', 'khaled '],
    'Age': [25, 22, 25, np.nan, 200],
    'City': ['Cairo', 'Dubai', 'Cairo', 'DUBAI', 'Riyadh'],
    'Salary': ['5000', '7000', '5000', 'missing', '4000']}
df=pd.DataFrame(data)
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df['City']=df['City'].str.strip().str.capitalize()
df
```

```python
df['Name']=df['Name'].str.strip()
df
```

```python
df['Age']=df['Age'].astype(int)
df
```

```python
df['Age']=df['Age'].replace(0, df['Age'].median())
df
```

```python
df.loc[df['Age'] > 100, 'Age']= df['Age'].median()
```

```python
df
```

```python
df['Salary']=pd.to_numeric(df['Salary'],errors='coerce')
df
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df
```

```python
df['Salary']=df['Salary'].apply(lambda x:"{:,.0f}".format(x))
df
```
